```python
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
iris=sn.load_dataset('iris')
```

```python
x=sn.PairGrid(iris)
x=x.map(plt.scatter)
```

```python
x=sn.PairGrid(iris,hue='species',palette='husl')
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```

```python
%matplotlib inline
x=sn.PairGrid(iris,hue='species',palette='husl')
x=x.map_diag(plt.hist,histtype='step',linewidth=2)
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```

```python
x=sn.PairGrid(iris,x_vars=['petal_length','petal_width'],y_vars=['sepal_length','sepal_width'])
x=x.map(plt.scatter)
```

```python

x=sn.PairGrid(iris)
x=x.map_diag(plt.hist)
x=x.map_upper(plt.scatter)
x=x.map_lower(sn.kdeplot)
```

```python
x=sn.PairGrid(iris,hue='species',hue_kws={'marker':['D','s','+']})
x=x.map(plt.scatter,s=40)
x=x.add_legend()
```
