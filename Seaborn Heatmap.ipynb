```python
import seaborn as sn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flights=sn.load_dataset('flights')
flights.head()
```

```python
flights=flights.pivot('month','year','passengers')
```

```python
sn.heatmap(flights,annot=True)
```

```python
sn.heatmap(flights,annot=True,fmt='d',vmin=0,vmax=500,center=flights.loc['September',1960])
```
