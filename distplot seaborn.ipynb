```python
sn.distplot(num)
```

```python
x_axis=pd.Series(num,name='x_axis')
sn.distplot(x_axis)
```

```python
y=sn.distplot(x_axis,vertical=True)
y
```

```python
sn.distplot(x_axis,hist=False)
```
