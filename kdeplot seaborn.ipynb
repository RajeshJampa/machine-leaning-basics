```python
import seaborn as sn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris=sn.load_dataset('iris')
```

```python
setosa=iris.loc[iris.species=='setosa']
versicolor=iris.loc[iris.species=='versicolor']
```

```python
sn.kdeplot(setosa.sepal_length,shade=True)
```

```python
sn.kdeplot(setosa.sepal_length,setosa.sepal_width,cmap='gist_rainbow')
```

```python
sn.kdeplot(versicolor.sepal_length,versicolor.sepal_width,cmap='gnuplot')
```

```python
sn.kdeplot(setosa.sepal_length,setosa.sepal_width,cmap='gist_rainbow')
sn.kdeplot(versicolor.sepal_length,versicolor.sepal_width,cmap='gnuplot')
```
